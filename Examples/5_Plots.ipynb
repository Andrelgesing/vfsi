```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
freq2d, phi2d = np.load('Data/phi_2d.npy')
freqSader, phiSader = np.load('Data/phi_sader.npy')
freq3d_cpu, phi3d_cpu = np.load('Data/phi_3d_uniform_cpu.npy')
freq3d_gpu, phi3d_gpu = np.load('Data/phi_3d_uniform_gpu.npy')
freq3d_gpu_non, phi3d_gpu_non = np.load('Data/phi_3d_nonuniform_gpu.npy')
freq3d_cpu_non, phi3d_cpu_non = np.load('Data/phi_3d_nonuniform_cpu.npy')
freq3d_cpu_non_1d, phi3d_cpu_non_1d = np.load('Data/phi_3d_nonuniform_cpu_1D.npy')
```

```python
plt.figure(1, figsize=(9, 4))
plt.semilogy(freq3d_cpu_non, phi3d_cpu_non,  linewidth=2.5, color='Gray', label='3D - CPU - non-uniform')
plt.plot(freq3d_gpu_non, phi3d_gpu_non, ':',  linewidth=2.5, color='Green', label='3D - GPU - non-uniform')
#plt.plot(freq3d_cpu, phi3d_cpu, color='DarkOrange', linewidth=2.4, label='3D - CPU - Uniform  ')

#plt.plot(freq3d_gpu, phi3d_gpu, color='Blue', linewidth=2.4, label='3D - GPU - Uniform')
plt.plot(freq3d_cpu_non_1d, phi3d_cpu_non_1d, color='red', linewidth=2.4, label='3D - CPU - Non-Uniform 1D ')

plt.plot(freq2d, phi2d, linewidth=2.4, color='Black', label='2D')
#plt.loglog(frequency/1e3, np.abs(phi_probe), '--', label='Old')
#plt.loglog(freqSader, phiSader, ':', linewidth=3,  label='Sader')

plt.legend(fontsize=12)
plt.xlabel('$f$ [kHz]', fontsize=14)
plt.ylabel(r'$\phi$ [nm/Pa]', fontsize=14)
plt.savefig('Figures/Spectrum_validation.png', dpi=400)
plt.show()
```
